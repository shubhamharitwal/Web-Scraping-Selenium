```python
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options

# Step 1: Get the webpage
url = "https://www.motor1.com/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Step 2: Type in the search bar '50 most expensive cars' using Selenium
options = Options()
options.headless = True  # Run Chrome in headless mode so that it doesn't open a browser window
driver = webdriver.Chrome(options=options)
driver.get(url)

search_input = driver.find_element("class name", "header-search__input")  # Corrected the method to find_element
search_input.send_keys("50 most expensive cars")
search_input.send_keys(Keys.RETURN)

# Wait for page to load completely
time.sleep(5)  # Adjust the waiting time as needed

# Step 4: Scrap the mentioned data and create the dataframe
soup = BeautifulSoup(driver.page_source, "html.parser")

cars_data = []
cars_container = soup.find("div", {"class": "search-results"})

if cars_container:
    cars_list = cars_container.find_all("div", {"class": "search-item"})
    for car in cars_list:
        car_name = car.find("h3", {"class": "search-result-title"}).text.strip()
        car_price = car.find("div", {"class": "search-result-price"}).text.strip()
        cars_data.append({"Car Name": car_name, "Price": car_price})

# Create DataFrame
df = pd.DataFrame(cars_data)
print(df)

# Close the webdriver
driver.quit()
```
